## Bloque 1. Importar librerías

En este bloque vamos a importar las librerías necesarias para:
- manejar datos numéricos,
- visualizar resultados,
- cargar un dataset de clasificación,
- entrenar un modelo base,
- y calcular métricas supervisadas.


```python
# Importamos NumPy para arreglos y operaciones numéricas.
import numpy as np

# Importamos Matplotlib para crear gráficas.
import matplotlib.pyplot as plt

# Importamos el dataset breast cancer.
from sklearn.datasets import load_breast_cancer

# Importamos la función para dividir los datos.
from sklearn.model_selection import train_test_split

# Importamos la regresión logística como modelo base.
from sklearn.linear_model import LogisticRegression

# Importamos métricas supervisadas.
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    roc_auc_score
)
```

### Interpretación del bloque 1

Este bloque prepara todo el entorno de trabajo.

**Qué significa:**
- Vamos a entrenar un modelo sencillo y luego analizarlo desde el punto de vista de la evaluación.
- El foco principal de esta práctica no es el algoritmo en sí, sino las métricas que nos permiten juzgar su desempeño.


## Bloque 2. Cargar el dataset y explorar su estructura

En este bloque vamos a:
- cargar un dataset real de clasificación,
- separar variables y clases,
- revisar dimensiones,
- y entender qué representa el problema.


```python
# Cargamos el dataset de cáncer de mama.
data = load_breast_cancer()

# Guardamos las variables de entrada en X.
X = data.data

# Guardamos la variable objetivo en y.
y = data.target

# Mostramos la forma de X.
print("Shape de X:", X.shape)

# Mostramos la forma de y.
print("Shape de y:", y.shape)

# Mostramos los nombres de las clases.
print("Clases:", data.target_names)

# Mostramos las primeras cinco observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos las primeras cinco etiquetas.
print("\nPrimeras 5 etiquetas de y:")
print(y[:5])
```

### Interpretación del bloque 2

La salida muestra que:
- hay varias observaciones clínicas,
- cada una está descrita por múltiples variables numéricas,
- y la variable objetivo es binaria.

**Qué significa pedagógicamente:**
- Este es un problema de clasificación binaria.
- Es ideal para estudiar métricas porque permite distinguir claramente entre positivos y negativos.


## Bloque 3. Dividir el dataset en entrenamiento y prueba

En este bloque vamos a separar los datos para entrenar el modelo con una parte y evaluarlo con otra.


```python
# Dividimos los datos en entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Variables de entrada
    y,                  # Clases
    test_size=0.3,      # 30% para prueba
    random_state=42,    # Reproducibilidad
    stratify=y          # Mantener proporción de clases
)

# Mostramos dimensiones del conjunto de entrenamiento.
print("Shape de X_train:", X_train.shape)

# Mostramos dimensiones del conjunto de prueba.
print("Shape de X_test:", X_test.shape)

# Mostramos dimensiones de y_train.
print("Shape de y_train:", y_train.shape)

# Mostramos dimensiones de y_test.
print("Shape de y_test:", y_test.shape)
```

### Interpretación del bloque 3

Con esta división el modelo se entrena en una parte del dataset y se evalúa en otra.

**Por qué es importante:**
- Permite medir qué tan bien generaliza el modelo.
- Evita evaluar en datos que ya vio durante el entrenamiento.


## Bloque 4. Entrenar un modelo base de clasificación

En este bloque vamos a usar regresión logística como modelo base para generar predicciones y probabilidades.


```python
# Creamos el modelo de regresión logística.
model = LogisticRegression(
    max_iter=1000,      # Aumentamos el número máximo de iteraciones
    random_state=42     # Reproducibilidad
)

# Entrenamos el modelo.
model.fit(X_train, y_train)

# Predecimos las clases del conjunto de prueba.
y_pred = model.predict(X_test)

# Calculamos las probabilidades estimadas para la clase positiva.
y_prob = model.predict_proba(X_test)[:, 1]

# Confirmamos que el modelo está listo.
print("Modelo entrenado correctamente")
```

### Interpretación del bloque 4

En este punto ya tenemos:
- predicciones de clase (`y_pred`)
- probabilidades (`y_prob`)

**Qué significa esto:**
- `y_pred` servirá para calcular accuracy, precision, recall y F1-score.
- `y_prob` será útil para construir la curva ROC y calcular AUC.


## Bloque 5. Calcular accuracy, precision, recall y F1-score

En este bloque vamos a calcular cuatro métricas clave para evaluar clasificación supervisada.


```python
# Calculamos accuracy.
acc = accuracy_score(y_test, y_pred)

# Calculamos precision.
prec = precision_score(y_test, y_pred)

# Calculamos recall.
rec = recall_score(y_test, y_pred)

# Calculamos F1-score.
f1 = f1_score(y_test, y_pred)

# Mostramos las métricas.
print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1-score:", f1)
```

### Interpretación del bloque 5

Estas métricas responden preguntas diferentes:
- **Accuracy**: ¿qué porcentaje total se clasificó correctamente?
- **Precision**: cuando el modelo dijo “positivo”, ¿cuántas veces acertó?
- **Recall**: de todos los positivos reales, ¿cuántos detectó?
- **F1-score**: balance entre precision y recall.

**Idea clave:**
Un modelo puede tener buena accuracy y aun así fallar en precision o recall, especialmente en datasets desbalanceados.


## Bloque 6. Construir la matriz de confusión y el reporte de clasificación

En este bloque vamos a mirar con más detalle en qué acierta y en qué falla el modelo.


```python
# Calculamos la matriz de confusión.
cm = confusion_matrix(y_test, y_pred)

# Generamos el reporte de clasificación.
report = classification_report(y_test, y_pred, target_names=data.target_names)

# Mostramos la matriz de confusión.
print("Matriz de confusión:")
print(cm)

# Mostramos el reporte completo.
print("\nReporte de clasificación:")
print(report)
```

### Interpretación del bloque 6

La matriz de confusión organiza los resultados en cuatro casos:
- verdaderos positivos
- verdaderos negativos
- falsos positivos
- falsos negativos

**Qué aporta esto:**
- Permite ver el tipo de error que está cometiendo el modelo.
- El reporte de clasificación resume precisión, recall y F1 por clase, lo que da una visión más completa que accuracy sola.


## Bloque 7. Visualizar la matriz de confusión

En este bloque vamos a convertir la matriz de confusión en una gráfica para interpretarla visualmente.


```python
# Creamos una figura para visualizar la matriz de confusión.
_, ax = plt.subplots(figsize=(6, 5))

# Dibujamos la matriz como imagen.
im = ax.imshow(cm, cmap="Blues")

# Agregamos una barra de color.
plt.colorbar(im)

# Definimos etiquetas de los ejes.
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(data.target_names)
ax.set_yticklabels(data.target_names)

# Etiquetamos los ejes.
ax.set_xlabel("Predicción")
ax.set_ylabel("Valor real")
ax.set_title("Matriz de confusión")

# Escribimos los valores dentro de cada celda.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha="center", va="center", color="black")

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 7

La visualización facilita la lectura de la matriz.

**Qué debe observarse:**
- La diagonal principal representa los aciertos.
- Las celdas fuera de la diagonal representan errores.
- Si la diagonal domina claramente, el modelo está funcionando bien.

**Mensaje docente:**
Esta gráfica ayuda mucho más que una tabla cuando se quiere explicar errores en clase.


## Bloque 8. Construir la curva ROC y calcular el AUC

En este bloque vamos a:
- construir la curva ROC,
- calcular el área bajo la curva (AUC),
- y visualizar la capacidad discriminativa del modelo.


```python
# Calculamos los puntos de la curva ROC.
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Calculamos el área bajo la curva ROC.
auc_score = roc_auc_score(y_test, y_prob)

# Mostramos el valor de AUC.
print("AUC-ROC:", auc_score)

# Creamos la figura.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la curva ROC.
ax.plot(fpr, tpr, label=f"ROC (AUC = {auc_score:.4f})")

# Dibujamos la diagonal de referencia.
ax.plot([0, 1], [0, 1], linestyle="--", color="gray")

# Agregamos etiquetas y título.
ax.set_xlabel("Tasa de falsos positivos")
ax.set_ylabel("Tasa de verdaderos positivos")
ax.set_title("Curva ROC")
ax.legend()

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 8

La curva ROC muestra cómo cambia la capacidad de detección del modelo según el umbral de decisión.

**Qué significa AUC:**
- Si AUC está cerca de 1, el modelo discrimina muy bien entre clases.
- Si AUC está cerca de 0.5, el modelo se comporta casi como azar.

**Valor pedagógico:**
La curva ROC permite evaluar el modelo sin depender de un único umbral fijo.


## Bloque 9. Comparar lo que mide cada métrica

En este bloque vamos a resumir en una tabla sencilla los valores calculados para discutirlos en conjunto.


```python
# Creamos un diccionario con las métricas calculadas.
metricas = {
    "Accuracy": acc,
    "Precision": prec,
    "Recall": rec,
    "F1-score": f1,
    "AUC-ROC": auc_score
}

# Recorremos el diccionario para mostrar las métricas.
for nombre, valor in metricas.items():
    print(f"{nombre}: {valor:.4f}")
```

### Interpretación del bloque 9

Este bloque permite comparar todas las métricas al mismo tiempo.

**Conclusión importante:**
- No existe una única métrica universalmente mejor.
- La métrica correcta depende del contexto.
- En problemas médicos o de fraude, recall y precision suelen ser más informativas que accuracy sola.


## Bloque 10. Hacer una predicción nueva y analizarla

En este bloque vamos a tomar una observación nueva y revisar tanto la clase predicha como la probabilidad asociada.


```python
# Tomamos la primera observación del conjunto de prueba como ejemplo.
nuevo_caso = X_test[0].reshape(1, -1)

# Predecimos la clase del nuevo caso.
pred_nuevo = model.predict(nuevo_caso)[0]

# Calculamos la probabilidad de la clase positiva.
prob_nuevo = model.predict_proba(nuevo_caso)[0]

# Mostramos el caso.
print("Nuevo caso:")
print(nuevo_caso)

# Mostramos la clase predicha.
print("\nClase predicha:", data.target_names[pred_nuevo])

# Mostramos las probabilidades por clase.
print("Probabilidades por clase:")
for i, prob in enumerate(prob_nuevo):
    print(f"  {data.target_names[i]}: {prob:.4f}")
```

### Interpretación del bloque 10

Este bloque conecta las métricas con una decisión concreta.

**Qué significa:**
- El modelo no solo entrega una clase, también entrega una probabilidad.
- Esa probabilidad ayuda a entender qué tan segura fue la predicción.
- En aplicaciones reales, esta información puede ser crucial para decidir acciones.
