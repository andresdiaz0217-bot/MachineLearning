## Objetivo
Comprender y aplicar el algoritmo **K-Means** para agrupar datos no etiquetados y analizar cómo se forman los clusters.


## Bloque 1. Importar librerías

En este bloque vamos a importar las librerías necesarias para:
- manejar arreglos numéricos,
- crear visualizaciones,
- generar un dataset sintético,
- entrenar el modelo K-Means,
- y evaluar la calidad del agrupamiento.


```python
# Importamos NumPy para trabajar con arreglos y operaciones numéricas.
import numpy as np

# Importamos Matplotlib para crear gráficas.
import matplotlib.pyplot as plt

# Importamos una función para generar datos sintéticos agrupados.
from sklearn.datasets import make_blobs

# Importamos el algoritmo K-Means.
from sklearn.cluster import KMeans

# Importamos métricas de evaluación de clustering.
from sklearn.metrics import silhouette_score
```

### Interpretación del bloque 1

Este bloque deja preparado el entorno de trabajo.

**Qué significa esto:**
- Ya están listas las herramientas para crear datos, entrenar el algoritmo y visualizar los resultados.
- A diferencia de clasificación supervisada, aquí no vamos a trabajar con etiquetas para entrenar el modelo, sino con agrupamiento no supervisado.


## Bloque 2. Generar un dataset sintético

En este bloque vamos a:
- crear un conjunto de datos artificial,
- simular varios grupos naturales,
- y observar la forma general del problema antes de entrenar el algoritmo.


```python
# Generamos un dataset sintético con 4 grupos.
X, y_real = make_blobs(
    n_samples=400,      # Número total de observaciones
    centers=4,          # Número real de grupos
    cluster_std=1.2,    # Dispersión de cada grupo
    random_state=42     # Semilla para reproducibilidad
)

# Mostramos el tamaño de la matriz de entrada.
print("Shape de X:", X.shape)

# Mostramos el tamaño del vector de clases reales.
print("Shape de y_real:", y_real.shape)

# Mostramos las primeras cinco observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos las primeras cinco etiquetas reales.
print("\nPrimeras 5 etiquetas reales:")
print(y_real[:5])
```

### Interpretación del bloque 2

La salida muestra que el dataset tiene:
- **400 observaciones**,
- **2 variables** por cada observación,
- y un patrón de agrupamiento real de **4 grupos**.

**Mensaje pedagógico:**
Aunque aquí existe `y_real`, K-Means no la usará para aprender.  
La incluimos solo como referencia pedagógica, para luego comparar lo que descubre el algoritmo.


## Bloque 3. Visualizar los datos originales

En este bloque vamos a representar gráficamente el dataset para observar la estructura visual de los grupos antes de aplicar K-Means.


```python
# Creamos una figura y un sistema de ejes.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos los datos usando los colores reales solo como referencia visual.
ax.scatter(
    X[:, 0],            # Primera variable en el eje X
    X[:, 1],            # Segunda variable en el eje Y
    c=y_real,           # Color según el grupo real
    cmap="viridis",     # Paleta de colores
    edgecolors="k"      # Borde negro para distinguir los puntos
)

# Agregamos un título a la gráfica.
ax.set_title("Datos originales con grupos reales")

# Etiquetamos los ejes.
ax.set_xlabel("Variable 1")
ax.set_ylabel("Variable 2")

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 3

La gráfica sugiere que existen varios grupos naturales en los datos.

**Qué debe observarse:**
- Algunos grupos están claramente separados.
- El problema parece apropiado para K-Means porque hay estructuras compactas y relativamente redondeadas.
- Esta visualización ayuda a entender por qué el algoritmo podría funcionar bien en este caso.


## Bloque 4. Entrenar K-Means con un número de clusters conocido

En este bloque vamos a:
- crear el modelo K-Means,
- indicar el número de clusters,
- entrenarlo con los datos,
- y obtener las etiquetas predichas para cada observación.


```python
# Definimos el número de clusters que queremos encontrar.
k = 4

# Creamos el modelo K-Means.
kmeans = KMeans(
    n_clusters=k,       # Número de clusters
    init="k-means++",   # Inicialización inteligente de centroides
    n_init=10,          # Número de inicializaciones
    random_state=42     # Reproducibilidad
)

# Entrenamos el modelo con los datos.
kmeans.fit(X)

# Guardamos la etiqueta de cluster asignada a cada observación.
clusters = kmeans.labels_

# Guardamos las coordenadas de los centroides finales.
centroides = kmeans.cluster_centers_

# Mostramos las primeras 10 etiquetas de cluster.
print("Primeras 10 etiquetas predichas:")
print(clusters[:10])

# Mostramos los centroides encontrados.
print("\nCentroides encontrados:")
print(centroides)
```

### Interpretación del bloque 4

En este bloque el algoritmo ya agrupó las observaciones.

**Qué significa esto:**
- Cada punto fue asignado a uno de los 4 clusters.
- Los centroides representan el “centro” de cada grupo encontrado.
- K-Means trabaja buscando minimizar la distancia entre los puntos y el centroide de su cluster.

**Mensaje conceptual:**
A diferencia de clasificación supervisada, aquí el modelo no predice clases reales; construye sus propios grupos basándose en similitud.


## Bloque 5. Visualizar los clusters encontrados y sus centroides

En este bloque vamos a:
- dibujar los clusters predichos por K-Means,
- mostrar los centroides,
- y comparar visualmente el agrupamiento con la distribución original.


```python
# Creamos una figura para mostrar el resultado del clustering.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos los puntos coloreados según el cluster asignado por K-Means.
ax.scatter(
    X[:, 0],            # Primera variable
    X[:, 1],            # Segunda variable
    c=clusters,         # Color según cluster encontrado
    cmap="viridis",     # Paleta de colores
    edgecolors="k"      # Borde negro
)

# Dibujamos los centroides como puntos rojos grandes.
ax.scatter(
    centroides[:, 0],   # Coordenadas X de los centroides
    centroides[:, 1],   # Coordenadas Y de los centroides
    c="red",            # Color rojo
    s=250,              # Tamaño grande
    marker="X",         # Forma en X
    label="Centroides"  # Etiqueta para la leyenda
)

# Agregamos título y nombres de ejes.
ax.set_title("Clusters encontrados por K-Means")
ax.set_xlabel("Variable 1")
ax.set_ylabel("Variable 2")

# Mostramos la leyenda.
ax.legend()

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 5

La gráfica muestra cómo K-Means organizó los datos.

**Qué se observa:**
- Los puntos quedaron agrupados según cercanía a los centroides.
- Los centroides se ubican en el centro aproximado de cada grupo.
- Si los clusters se parecen mucho a la estructura visual original, el modelo está haciendo un buen trabajo.

**Idea clave:**
K-Means no “conoce” los grupos reales, pero puede aproximarlos si los datos tienen una estructura adecuada.


## Bloque 6. Evaluar la calidad del agrupamiento

En este bloque vamos a usar dos medidas:
- **Inercia**: qué tan cerca están los puntos de su centroide.
- **Silhouette Score**: qué tan bien separados están los clusters entre sí.


```python
# Calculamos la inercia del modelo.
inercia = kmeans.inertia_

# Calculamos el índice silhouette.
silhouette = silhouette_score(X, clusters)

# Mostramos la inercia.
print("Inercia del modelo:", inercia)

# Mostramos el silhouette score.
print("Silhouette Score:", silhouette)
```

### Interpretación del bloque 6

Estas métricas ayudan a evaluar el agrupamiento.

**Inercia:**
- Mide qué tan compactos son los clusters.
- Mientras menor sea, más cerca están los puntos de su centroide.

**Silhouette Score:**
- Toma valores entre -1 y 1.
- Valores cercanos a 1 indican clusters bien definidos y separados.
- Valores cercanos a 0 indican superposición.
- Valores negativos indican mala asignación.

**Conclusión docente:**
No basta con mirar la gráfica; también es importante respaldar el análisis con métricas.


## Bloque 7. Aplicar el método del codo

En este bloque vamos a probar varios valores de `k` para observar cómo cambia la inercia.

El objetivo es identificar un punto donde aumentar el número de clusters ya no mejora tanto el resultado.


```python
# Creamos una lista para guardar la inercia de cada valor de k.
inercias = []

# Definimos el rango de valores de k que vamos a probar.
rango_k = range(1, 11)

# Recorremos cada valor de k.
for k in rango_k:
    # Creamos un modelo K-Means con ese valor de k.
    model = KMeans(
        n_clusters=k,       # Número de clusters
        init="k-means++",   # Inicialización inteligente
        n_init=10,          # Número de inicializaciones
        random_state=42     # Reproducibilidad
    )

    # Entrenamos el modelo.
    model.fit(X)

    # Guardamos la inercia.
    inercias.append(model.inertia_)

# Creamos la figura del método del codo.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la curva de inercia.
ax.plot(rango_k, inercias, marker="o")

# Agregamos título y nombres de ejes.
ax.set_title("Método del codo para elegir k")
ax.set_xlabel("Número de clusters (k)")
ax.set_ylabel("Inercia")

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 7

La curva del codo ayuda a elegir un valor razonable de `k`.

**Qué debe buscarse:**
- Un punto donde la curva deja de caer bruscamente.
- Ese “quiebre” sugiere que aumentar más los clusters aporta menos mejora.

**Lectura típica:**
Si el codo aparece cerca de 4, eso confirma que elegir `k = 4` fue razonable para este dataset.


## Bloque 8. Predecir el cluster de nuevos puntos

En este bloque vamos a crear nuevas observaciones ficticias y pedirle al modelo que indique a qué cluster pertenecerían.


```python
# Creamos nuevos puntos ficticios.
nuevos_puntos = np.array([
    [4.5, 2.0],    # Punto 1
    [-6.0, -7.0],  # Punto 2
    [0.0, 8.0]     # Punto 3
])

# Predecimos el cluster al que pertenece cada nuevo punto.
pred_nuevos = kmeans.predict(nuevos_puntos)

# Mostramos los nuevos puntos.
print("Nuevos puntos:")
print(nuevos_puntos)

# Mostramos el cluster asignado a cada uno.
print("\nClusters predichos:")
print(pred_nuevos)
```

### Interpretación del bloque 8

Este bloque muestra que K-Means también puede usarse para asignar nuevos datos a grupos ya aprendidos.

**Qué significa:**
- Cada nuevo punto se asigna al centroide más cercano.
- Esto permite usar K-Means no solo para explorar datos, sino también para clasificar nuevas observaciones dentro de grupos descubiertos previamente.
