# Taller resuelto: Kernels en SVM

Este notebook sigue la misma estructura pedagógica:

1. **Antes de cada bloque de código** se explica qué se va a hacer.  
2. **Dentro del código** cada línea contiene comentarios.

## Objetivo
Comprender cómo cambian los resultados de **SVM** cuando usamos distintos **kernels**: lineal, polinómico y RBF.


## Bloque 1. Importar librerías

En este bloque vamos a importar las librerías necesarias para:

- generar un dataset no lineal,
- visualizarlo,
- dividir los datos,
- entrenar varios modelos SVM con kernels diferentes,
- y evaluar su comportamiento.


```python
# Importamos NumPy para arreglos y operaciones numéricas.
import numpy as np

# Importamos Matplotlib para crear gráficas.
import matplotlib.pyplot as plt

# Importamos una función para generar un dataset no lineal.
from sklearn.datasets import make_moons

# Importamos la función para dividir los datos.
from sklearn.model_selection import train_test_split

# Importamos el clasificador SVM.
from sklearn.svm import SVC

# Importamos métricas para evaluar el modelo.
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Interpretación del bloque 1

Este bloque prepara el entorno de trabajo para estudiar kernels.

**Punto importante:**
- Aquí ya no usaremos Iris sino un dataset no lineal, ideal para mostrar por qué los kernels son importantes.


## Bloque 2. Generar un dataset no lineal

En este bloque vamos a:

- crear un dataset con forma de lunas,
- revisar su tamaño,
- y preparar un problema donde comparar kernels tenga sentido.


```python
# Generamos un dataset sintético con forma de lunas.
X, y = make_moons(
    n_samples=300,      # Número total de observaciones
    noise=0.25,         # Cantidad de ruido
    random_state=42     # Semilla para reproducibilidad
)

# Mostramos el tamaño de la matriz de entrada.
print("Shape de X:", X.shape)

# Mostramos el tamaño del vector de clases.
print("Shape de y:", y.shape)

# Mostramos las primeras cinco observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos las primeras cinco etiquetas.
print("\nPrimeras 5 etiquetas de y:")
print(y[:5])
```

### Interpretación del bloque 2

La salida muestra que el dataset tiene:

- **300 observaciones**,
- **2 variables**,
- y **2 clases**.

**Qué significa didácticamente:**
- Es un problema de clasificación binaria.
- El dataset fue diseñado para que una separación lineal no sea perfecta.
- Esto lo hace ideal para estudiar por qué existen los kernels.


## Bloque 3. Visualizar el dataset

En este bloque vamos a dibujar el dataset para ver su forma y entender por qué ciertos kernels pueden funcionar mejor que otros.


```python
# Creamos una figura y un sistema de ejes.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la dispersión de las observaciones.
ax.scatter(
    X[:, 0],            # Primera variable en el eje X
    X[:, 1],            # Segunda variable en el eje Y
    c=y,                # Color según clase
    cmap="coolwarm",    # Paleta de colores
    edgecolors="k"      # Borde negro
)

# Agregamos título y etiquetas.
ax.set_title("Dataset sintético: make_moons")
ax.set_xlabel("Variable 1")
ax.set_ylabel("Variable 2")

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 3

La gráfica muestra una estructura claramente no lineal.

**Qué se observa:**
- Las dos clases tienen forma de lunas entrelazadas.
- No parece fácil separarlas con una línea recta.
- Esto anticipa que un kernel lineal puede tener limitaciones.

**Valor pedagógico:**
Aquí se entiende visualmente por qué SVM necesita kernels no lineales en ciertos problemas.


## Bloque 4. Dividir los datos en entrenamiento y prueba

En este bloque se separan los datos para entrenar y luego evaluar los modelos con kernels distintos.


```python
# Dividimos el dataset en entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Variables de entrada
    y,                  # Clases
    test_size=0.3,      # 30% para prueba
    random_state=42,    # Reproducibilidad
    stratify=y          # Mantener proporción de clases
)

# Mostramos dimensiones del entrenamiento.
print("Shape de X_train:", X_train.shape)

# Mostramos dimensiones de prueba.
print("Shape de X_test:", X_test.shape)

# Mostramos dimensiones de las etiquetas de entrenamiento.
print("Shape de y_train:", y_train.shape)

# Mostramos dimensiones de las etiquetas de prueba.
print("Shape de y_test:", y_test.shape)
```

### Interpretación del bloque 4

El dataset queda dividido en dos partes:

- una para aprender,
- otra para evaluar.

Esto garantiza una comparación más justa entre kernels porque todos se prueban sobre datos no vistos.


## Bloque 5. Entrenar tres modelos SVM con kernels diferentes

En este bloque vamos a entrenar:

- un SVM lineal,
- un SVM con kernel polinómico,
- un SVM con kernel RBF.

Luego compararemos sus resultados.


```python
# Creamos el modelo con kernel lineal.
svm_linear = SVC(
    kernel="linear",    # Kernel lineal
    C=1.0,              # Regularización
    random_state=42     # Reproducibilidad
)

# Creamos el modelo con kernel polinómico.
svm_poly = SVC(
    kernel="poly",      # Kernel polinómico
    degree=3,           # Grado del polinomio
    C=1.0,              # Regularización
    gamma="scale",      # Escalado automático
    random_state=42     # Reproducibilidad
)

# Creamos el modelo con kernel RBF.
svm_rbf = SVC(
    kernel="rbf",       # Kernel de base radial
    C=1.0,              # Regularización
    gamma="scale",      # Escalado automático
    random_state=42     # Reproducibilidad
)

# Entrenamos el modelo lineal.
svm_linear.fit(X_train, y_train)

# Entrenamos el modelo polinómico.
svm_poly.fit(X_train, y_train)

# Entrenamos el modelo RBF.
svm_rbf.fit(X_train, y_train)

# Confirmamos el entrenamiento.
print("Modelos SVM entrenados correctamente")
```

### Interpretación del bloque 5

En este punto ya tenemos tres modelos distintos sobre el mismo problema.

**Qué significa esto:**
- Todos usan SVM, pero cambian la forma de construir la frontera de decisión.
- El kernel lineal buscará una separación recta.
- El kernel polinómico permitirá curvaturas.
- El kernel RBF generará fronteras más flexibles y locales.

**Idea docente clave:**
El kernel no cambia el objetivo del modelo, pero sí cambia la forma en que entiende el espacio.


## Bloque 6. Evaluar los tres modelos

En este bloque vamos a:

- predecir con cada kernel,
- calcular accuracy,
- y comparar el desempeño obtenido.


```python
# Generamos predicciones con el kernel lineal.
pred_linear = svm_linear.predict(X_test)

# Generamos predicciones con el kernel polinómico.
pred_poly = svm_poly.predict(X_test)

# Generamos predicciones con el kernel RBF.
pred_rbf = svm_rbf.predict(X_test)

# Calculamos accuracy para el kernel lineal.
acc_linear = accuracy_score(y_test, pred_linear)

# Calculamos accuracy para el kernel polinómico.
acc_poly = accuracy_score(y_test, pred_poly)

# Calculamos accuracy para el kernel RBF.
acc_rbf = accuracy_score(y_test, pred_rbf)

# Mostramos los resultados.
print("Accuracy kernel lineal:", acc_linear)
print("Accuracy kernel polinómico:", acc_poly)
print("Accuracy kernel RBF:", acc_rbf)
```

### Interpretación del bloque 6

Aquí se comparan los resultados cuantitativos de los tres kernels.

**Qué suele ocurrir en este tipo de dataset:**
- El kernel lineal suele quedarse corto.
- El polinómico puede mejorar, pero depende mucho del grado.
- El RBF suele adaptarse mejor a formas curvas como las lunas.

**Valor conceptual:**
Este bloque demuestra que el kernel adecuado depende de la geometría del problema.


## Bloque 7. Revisar matriz de confusión y reporte del mejor modelo

En este bloque vamos a elegir el mejor kernel según accuracy y revisar con más detalle su rendimiento.


```python
# Seleccionamos el mejor modelo comparando accuracy.
best_name = max(
    [("lineal", acc_linear), ("polinómico", acc_poly), ("rbf", acc_rbf)],
    key=lambda x: x[1]
)[0]

# Elegimos el modelo ganador según su nombre.
best_model = svm_linear if best_name == "lineal" else svm_poly if best_name == "polinómico" else svm_rbf

# Elegimos también las predicciones del mejor modelo.
best_pred = pred_linear if best_name == "lineal" else pred_poly if best_name == "polinómico" else pred_rbf

# Calculamos la matriz de confusión.
cm = confusion_matrix(y_test, best_pred)

# Generamos el reporte de clasificación.
report = classification_report(y_test, best_pred)

# Mostramos el mejor kernel encontrado.
print("Mejor kernel:", best_name)

# Mostramos la matriz de confusión.
print("\nMatriz de confusión:")
print(cm)

# Mostramos el reporte de clasificación.
print("\nReporte de clasificación:")
print(report)
```

### Interpretación del bloque 7

Este bloque profundiza en el mejor modelo.

**Qué debe observarse:**
- La matriz de confusión muestra en qué clase se equivoca el modelo.
- El reporte permite revisar precisión, recall y F1-score.
- Si el mejor kernel tiene muy pocos errores, significa que logró representar bien la forma del problema.

**Lección clave:**
No basta con escoger un modelo por intuición; también hay que verificarlo con métricas.


## Bloque 8. Visualizar las fronteras de decisión de los tres kernels

En este bloque vamos a dibujar las regiones de decisión de:

- kernel lineal,
- kernel polinómico,
- kernel RBF.

Esto permitirá comparar visualmente cómo cambia la frontera según el kernel.


```python
# Definimos un margen para el plano.
offset = 0.5

# Calculamos los límites del eje X.
x_min, x_max = X[:, 0].min() - offset, X[:, 0].max() + offset

# Calculamos los límites del eje Y.
y_min, y_max = X[:, 1].min() - offset, X[:, 1].max() + offset

# Construimos la malla de puntos.
x_grid, y_grid = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Unimos las coordenadas de la malla en una sola matriz.
grid_points = np.c_[x_grid.ravel(), y_grid.ravel()]

# Predecimos las clases de la malla con el modelo lineal.
pred_linear_grid = svm_linear.predict(grid_points).reshape(x_grid.shape)

# Predecimos las clases de la malla con el modelo polinómico.
pred_poly_grid = svm_poly.predict(grid_points).reshape(x_grid.shape)

# Predecimos las clases de la malla con el modelo RBF.
pred_rbf_grid = svm_rbf.predict(grid_points).reshape(x_grid.shape)

# Creamos una figura con tres subgráficas.
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Dibujamos el fondo del kernel lineal.
axes[0].pcolormesh(x_grid, y_grid, pred_linear_grid, cmap="coolwarm", shading="auto")

# Superponemos los datos reales en la primera gráfica.
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")

# Agregamos título a la primera gráfica.
axes[0].set_title("Kernel lineal")

# Dibujamos el fondo del kernel polinómico.
axes[1].pcolormesh(x_grid, y_grid, pred_poly_grid, cmap="coolwarm", shading="auto")

# Superponemos los datos reales en la segunda gráfica.
axes[1].scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")

# Agregamos título a la segunda gráfica.
axes[1].set_title("Kernel polinómico")

# Dibujamos el fondo del kernel RBF.
axes[2].pcolormesh(x_grid, y_grid, pred_rbf_grid, cmap="coolwarm", shading="auto")

# Superponemos los datos reales en la tercera gráfica.
axes[2].scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")

# Agregamos título a la tercera gráfica.
axes[2].set_title("Kernel RBF")

# Ajustamos automáticamente los espacios de la figura.
plt.tight_layout()

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 8

Esta es la visualización más importante del notebook.

**Qué muestra:**
- El kernel lineal intenta separar con una frontera recta y por eso suele fallar.
- El kernel polinómico genera una frontera curva.
- El kernel RBF produce regiones más flexibles que suelen ajustarse mejor a este tipo de datos.

**Mensaje docente:**
Aquí se entiende claramente por qué existen los kernels: permiten que SVM resuelva problemas donde una línea recta no basta.


## Bloque 9. Hacer una predicción nueva con el mejor kernel

En este bloque vamos a:

- elegir el mejor kernel,
- crear una nueva observación,
- y consultar cuál clase predice el modelo.


```python
# Creamos una nueva observación ficticia con dos variables.
nuevo_punto = np.array([[1.2, 0.1]])

# Usamos el mejor modelo para predecir su clase.
pred_nuevo = best_model.predict(nuevo_punto)[0]

# Mostramos el nuevo punto.
print("Nuevo punto:", nuevo_punto)

# Mostramos la clase predicha.
print("Clase predicha:", pred_nuevo)
```

### Interpretación del bloque 9

Este bloque conecta la comparación de kernels con una decisión práctica.

**Qué significa:**
- Ya no estamos solo comparando gráficas y métricas.
- Ahora el mejor modelo se usa para clasificar una nueva observación.
- Esto muestra el objetivo final del aprendizaje supervisado: tomar decisiones sobre datos no vistos.


## Cierre del ejercicio de kernels en SVM

### Ideas clave
- El kernel lineal funciona bien cuando la separación puede aproximarse con rectas.
- El kernel polinómico añade curvatura.
- El kernel RBF suele ser muy potente en problemas no lineales.
- Elegir el kernel correcto depende de la forma del problema.

### Conclusión general
No existe un kernel universalmente mejor; su conveniencia depende de cómo se distribuyen los datos en el espacio.
