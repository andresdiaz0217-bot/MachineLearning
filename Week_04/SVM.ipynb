# Taller : Support Vector Machine (SVM)

## Objetivo
Comprender y aplicar un modelo **SVM lineal** para resolver un problema de clasificación usando el dataset **Iris**.


## Bloque 1. Importar librerías

En este bloque vamos a importar las librerías necesarias para:

- manipular arreglos numéricos,
- graficar datos,
- cargar el dataset Iris,
- dividir los datos en entrenamiento y prueba,
- entrenar un modelo SVM,
- y evaluar sus resultados.


```python
# Importamos NumPy para manejar arreglos y operaciones numéricas.
import numpy as np

# Importamos Matplotlib para crear visualizaciones.
import matplotlib.pyplot as plt

# Importamos el dataset Iris desde scikit-learn.
from sklearn.datasets import load_iris

# Importamos la función para dividir los datos en entrenamiento y prueba.
from sklearn.model_selection import train_test_split

# Importamos el clasificador SVM.
from sklearn.svm import SVC

# Importamos métricas para evaluar el desempeño del modelo.
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Interpretación del bloque 1

Este bloque deja preparado el entorno de trabajo.

**Qué significa esto:**
- Ya tenemos listas las herramientas para cargar datos, construir el modelo, visualizar el problema y evaluar resultados.
- Aunque todavía no hay un resultado de Machine Learning, este paso es fundamental porque define todo lo que podremos hacer después.


## Bloque 2. Cargar el dataset y explorar su estructura

En este bloque vamos a:

- cargar el dataset Iris,
- separar variables de entrada y clases,
- revisar dimensiones,
- y conocer las etiquetas y variables disponibles.


```python
# Cargamos el dataset Iris en memoria.
iris = load_iris()

# Guardamos la matriz de características en X.
X = iris.data

# Guardamos las etiquetas de clase en y.
y = iris.target

# Mostramos el tamaño de la matriz de entrada.
print("Shape de X:", X.shape)

# Mostramos el tamaño del vector de clases.
print("Shape de y:", y.shape)

# Mostramos los nombres de las clases.
print("Clases:", iris.target_names)

# Mostramos los nombres de las variables del dataset.
print("Variables:", iris.feature_names)

# Mostramos las primeras cinco observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos las primeras cinco clases.
print("\nPrimeras 5 etiquetas de y:")
print(y[:5])
```

### Interpretación del bloque 2

La salida muestra que:

- hay **150 observaciones**,
- cada observación tiene **4 variables**,
- y el problema tiene **3 clases**.

**Qué debemos entender:**
- Estamos frente a un problema de **clasificación supervisada**.
- Cada flor se representa con variables numéricas y una clase conocida.
- Iris sigue siendo un dataset ideal para enfocarse en la lógica del modelo más que en la limpieza de datos.


## Bloque 3. Visualizar dos variables para entender el problema

En este bloque vamos a seleccionar dos variables para representar visualmente el dataset en 2D.

Esto servirá para:
- observar la separación entre clases,
- generar intuición,
- y preparar la posterior visualización del modelo SVM.


```python
# Seleccionamos solo las dos primeras variables para poder graficar en 2D.
X_vis = X[:, 0:2]

# Creamos una figura y un sistema de ejes.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la dispersión de las observaciones.
ax.scatter(
    X_vis[:, 0],        # Primera variable en el eje X
    X_vis[:, 1],        # Segunda variable en el eje Y
    c=y,                # Color según la clase
    cmap='Set1',        # Paleta de colores
    edgecolors='k'      # Borde negro para distinguir los puntos
)

# Agregamos título a la gráfica.
ax.set_title("Distribución del dataset Iris (2 variables)")

# Asignamos nombre al eje X.
ax.set_xlabel(iris.feature_names[0])

# Asignamos nombre al eje Y.
ax.set_ylabel(iris.feature_names[1])

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 3

La gráfica permite ver cómo se distribuyen las clases cuando usamos solo dos variables.

**Qué puede observarse:**
- Algunas clases parecen más separadas.
- Otras pueden mostrar cierto solapamiento.
- Esto sugiere qué tan difícil puede ser trazar un límite de decisión.

**Valor pedagógico:**
Antes de entrenar un modelo, ver los datos ayuda a construir intuición sobre la geometría del problema.


## Bloque 4. Dividir los datos en entrenamiento y prueba

En este bloque vamos a separar el dataset en dos subconjuntos:

- uno para entrenar,
- otro para probar.

Esto permite evaluar el modelo con observaciones que no vio durante el aprendizaje.


```python
# Dividimos los datos en entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Variables de entrada
    y,                  # Clases
    test_size=0.3,      # 30% de los datos se reservan para prueba
    random_state=42,    # Semilla para reproducibilidad
    stratify=y          # Mantiene la proporción de clases
)

# Mostramos la forma del conjunto de entrenamiento.
print("Shape de X_train:", X_train.shape)

# Mostramos la forma del conjunto de prueba.
print("Shape de X_test:", X_test.shape)

# Mostramos la forma de las etiquetas de entrenamiento.
print("Shape de y_train:", y_train.shape)

# Mostramos la forma de las etiquetas de prueba.
print("Shape de y_test:", y_test.shape)
```

### Interpretación del bloque 4

La salida confirma que el dataset quedó dividido correctamente.

**Qué significa:**
- El modelo aprenderá con el conjunto de entrenamiento.
- Después será evaluado con el conjunto de prueba.
- Esto evita medir el desempeño sobre los mismos datos usados para aprender, lo cual daría una visión poco realista.


## Bloque 5. Crear y entrenar el modelo SVM lineal

En este bloque vamos a:

- construir un modelo SVM,
- usar un kernel lineal,
- entrenarlo con el conjunto de entrenamiento,
- y dejarlo listo para hacer predicciones.


```python
# Creamos un modelo SVM con kernel lineal.
svm_model = SVC(
    kernel="linear",    # Elegimos un separador lineal
    C=1.0,              # Parámetro de regularización
    probability=True,   # Permite calcular probabilidades
    random_state=42     # Semilla para reproducibilidad
)

# Entrenamos el modelo con los datos de entrenamiento.
svm_model.fit(X_train, y_train)

# Confirmamos que el entrenamiento terminó correctamente.
print("Modelo SVM lineal entrenado correctamente")
```

### Interpretación del bloque 5

El modelo ya quedó entrenado.

**Qué significa usar un kernel lineal:**
- SVM intentará encontrar hiperplanos que separen las clases de la mejor manera posible.
- La idea central es maximizar el margen entre clases.

**Qué representa `C`:**
- Controla el equilibrio entre un margen amplio y la penalización por errores.
- Un valor moderado como `1.0` es un buen punto de partida.


## Bloque 6. Hacer predicciones y evaluar el modelo

En este bloque vamos a:

- predecir las clases del conjunto de prueba,
- calcular accuracy,
- ver la matriz de confusión,
- y revisar el reporte de clasificación.


```python
# Predecimos las clases del conjunto de prueba.
y_pred = svm_model.predict(X_test)

# Calculamos la exactitud del modelo.
acc = accuracy_score(y_test, y_pred)

# Calculamos la matriz de confusión.
cm = confusion_matrix(y_test, y_pred)

# Generamos el reporte de clasificación.
report = classification_report(y_test, y_pred, target_names=iris.target_names)

# Mostramos el accuracy.
print("Accuracy:", acc)

# Mostramos la matriz de confusión.
print("\nMatriz de confusión:")
print(cm)

# Mostramos el reporte completo.
print("\nReporte de clasificación:")
print(report)
```

### Interpretación del bloque 6

Este bloque muestra el desempeño real del modelo en datos no vistos.

**Cómo interpretar los resultados:**
- **Accuracy** muestra el porcentaje global de aciertos.
- **Matriz de confusión** permite ver en qué clases se equivoca el modelo.
- **Reporte de clasificación** resume precisión, recall y F1-score.

**Qué suele pasar con SVM en Iris:**
- El desempeño suele ser alto.
- Esto se debe a que Iris tiene una estructura relativamente favorable para la clasificación.


## Bloque 7. Visualizar las fronteras de decisión del SVM lineal

En este bloque vamos a:

- entrenar un SVM lineal usando solo dos variables,
- construir una malla en el plano,
- predecir la clase de cada punto,
- y visualizar las fronteras de decisión.


```python
# Creamos un modelo SVM lineal para trabajar solo en 2D.
svm_vis = SVC(
    kernel="linear",    # Kernel lineal
    C=1.0,              # Regularización
    random_state=42     # Reproducibilidad
)

# Entrenamos el modelo con las dos variables visuales.
svm_vis.fit(X_vis, y)

# Definimos un margen visual.
offset = 0.5

# Calculamos el rango del eje X.
x_min, x_max = X_vis[:, 0].min() - offset, X_vis[:, 0].max() + offset

# Calculamos el rango del eje Y.
y_min, y_max = X_vis[:, 1].min() - offset, X_vis[:, 1].max() + offset

# Construimos una malla de puntos.
x_grid, y_grid = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

# Predecimos la clase de cada punto de la malla.
pred_grid = svm_vis.predict(np.c_[x_grid.ravel(), y_grid.ravel()])

# Reorganizamos las predicciones con la forma de la malla.
pred_grid = pred_grid.reshape(x_grid.shape)

# Creamos la figura.
_, ax = plt.subplots(figsize=(7, 5))

# Pintamos el fondo con la clase predicha en cada región.
ax.pcolormesh(
    x_grid,
    y_grid,
    pred_grid,
    cmap="Set1",
    shading="auto"
)

# Superponemos los datos reales.
ax.scatter(
    X_vis[:, 0],
    X_vis[:, 1],
    c=y,
    cmap="Set1",
    edgecolors="k"
)

# Agregamos título y etiquetas.
ax.set_title("Fronteras de decisión con SVM lineal")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

# Mostramos la gráfica.
plt.tight_layout()
plt.show()
```

### Interpretación del bloque 7

Esta gráfica permite observar cómo SVM divide el plano entre clases.

**Qué suele verse:**
- Las fronteras son más limpias y regulares que en modelos basados en vecinos.
- Esto se debe a que SVM busca hiperplanos con máximo margen.
- El fondo muestra qué clase sería asignada a nuevas observaciones en cada región.

**Mensaje didáctico:**
SVM no solo clasifica puntos; construye una frontera óptima que separa clases con el mayor margen posible.


## Bloque 8. Hacer una predicción nueva

En este bloque vamos a crear una nueva flor ficticia y pedirle al modelo:

- la clase predicha,
- el nombre de la especie,
- y las probabilidades por clase.


```python
# Creamos una nueva flor con cuatro características.
nueva_flor = np.array([[5.8, 3.0, 4.35, 1.3]])

# Predecimos la clase de la nueva flor.
pred_nueva = svm_model.predict(nueva_flor)[0]

# Calculamos las probabilidades por clase.
prob_nueva = svm_model.predict_proba(nueva_flor)[0]

# Mostramos la observación.
print("Nueva flor:", nueva_flor)

# Mostramos la clase predicha en número.
print("Clase predicha:", pred_nueva)

# Mostramos el nombre real de la especie.
print("Nombre de la clase:", iris.target_names[pred_nueva])

# Mostramos las probabilidades por clase.
print("Probabilidades por clase:")
for i, prob in enumerate(prob_nueva):
    print(f"  {iris.target_names[i]}: {prob:.4f}")
```

### Interpretación del bloque 8

Este bloque conecta el modelo con una situación concreta.

**Cómo interpretar la salida:**
- La clase predicha corresponde a la región del espacio en la que cae la nueva flor.
- Las probabilidades ayudan a ver si la decisión fue clara o si hubo cierta competencia entre clases.
- Aquí el estudiante puede ver el paso final del flujo: **dato nuevo → modelo → decisión**.
